# Library Imports

```python
import pandas as pd
import numpy as np
import csv
```

# Reading Roast Dataset and Karma Score Filtering

```python
roast_df = pd.read_csv("./Raw/RoastMe.csv", sep="	", engine='python')
```

```python
roast_df = roast_df[roast_df['K_Score'] > 3]
```

# Creating Roast Comments NumPy Array

```python
roast_comments_arr = np.array(roast_df['Comment'])
```

# Tokenizing Sentences using NLTK

```python
from nltk.tokenize import sent_tokenize
```

```python
import nltk
#nltk.download()
```

```python
roast_tokenized_sentence = []
for i in range(0, 146848):
    s = sent_tokenize(roast_comments_arr[i])
    roast_tokenized_sentence.append(s)
for i in range(146849, len(roast_comments_arr)):
    s = sent_tokenize(roast_comments_arr[i])
    roast_tokenized_sentence.append(s)
```

# Lemma Checking & Attribute Based Filtering

```python
attr_list = []
file1 = open("attr-list.txt", "r")

x = file1.readlines()

for i in x:
    str1 = ""
    for k in i:
        if(k == ',' or k == '\n'):
            break
        else:
            str1+=k
    attr_list.append(str1)

print(attr_list)
```

## Lemmatization using SpaCy

```python
#!pip install -U spacy
#!python -m spacy download en_core_web_sm
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
len(roast_tokenized_sentence)
```

```python
final_roast_sent = []
for i in roast_tokenized_sentence:
    temp = []
    for k in i:
        sent = nlp(k)
        flag = False
        for tok in sent:
            if(tok.lemma_ in attr_list):
                flag = True
        if flag == True:
            temp.append(k)
        else:
            pass
    final_roast_sent.append(temp)

final_roast_sent
```

```python
len(final_roast_sent)
```

```python
f = []
for i in final_roast_sent:
    if(i != []):
        f.append(i)
```
